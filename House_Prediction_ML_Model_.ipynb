#linear regression practical implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import fetch_openml
boston = fetch_openml(name = 'boston', version = 1, as_frame = True, parser = 'pandas')
```

```python
boston
```

```python
df = boston.frame
```

```python
df.head()
```

```python
df.info()
```

```python
df['Price'] = boston.target
df.head()
```

## Dividing Depend and Independent into Feature

```python
x = df.iloc[ :,:-1]
```

```python
df.head()
```

```python
y = df.iloc[:,-1]
```

```python
y.head()
```

## Implement Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

x = x.astype(float)
y = y.astype(float)

lin_reg = LinearRegression()
mse = cross_val_score(lin_reg, x, y, scoring = 'neg_mean_squared_error', cv = 5)
print(mse)
lin_reg.fit(x,y)
lin_reg.score(x,y)
```

```python
mean_mse = np.mean(mse)
print("MSE scores per fold: ", mse, "\n")
print("Average Negative MSE: ", mean_mse )
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
lin_reg.fit(x_train, y_train)
```

```python
prediction = lin_reg.predict(x_test)
```

```python
print (prediction)
```

##Compare Few Value VS Actual Value

```python
import pandas as pd
comparison_df = pd.DataFrame({'Actual Value': y_test, 'Predicted Value': prediction})
print(comparison_df.head())
```

```python
from sklearn import metrics
import numpy as np

#1. R-squared Score (Best Possible Score is 1.0 )
r2 = metrics.r2_score(y_test, prediction)

#2. Mean Absolute Error  (Average Error in "Price" Units)
mae = metrics.mean_absolute_error(y_test, prediction)

#3. Root Mean Squared Error (Standard Division Of Prediction Errors)
rmse = np.sqrt(metrics.mean_squared_error(y_test, prediction))

print(f"R-squared Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}" )
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```
